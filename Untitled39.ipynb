#  Portfolio Return Analysis and Classification

This project analyzes whether a portfolio consisting of KYD Eurobond, BIST 100, and Gold outperforms a 1-month KYD deposit index.  
Additionally, it includes a binary classification task to predict whether the daily return will be positive.


```python
# 1. Load Data and Clean
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Simulate return data for 500 days
np.random.seed(42)
dates = pd.date_range(start="2020-01-01", periods=500, freq="D")

# Generate portfolio and benchmark daily returns
portfolio_returns = np.random.normal(0.001, 0.01, size=500)
benchmark_returns = np.random.normal(0.0008, 0.009, size=500)

# Create DataFrame
df = pd.DataFrame({
    "Portfolio": portfolio_returns,
    "Benchmark": benchmark_returns
}, index=dates)
```

## 2. Feature Engineering

I create:
- `Lag_1`: Yesterday's return
- `Volatility_5d`: 5-day rolling standard deviation
- `Direction`: Binary target (1 if positive return, else 0)


```python
df["Lag_1"] = df["Portfolio"].shift(1)
df["Volatility_5d"] = df["Portfolio"].rolling(5).std()
df["Direction"] = (df["Portfolio"] > 0).astype(int)
df.dropna(inplace=True)
```

## 3. Exploratory Data Analysis (EDA)

I visualize return trends, distribution, and normality:


```python
# Time series plot
plt.figure(figsize=(12, 4))
df["Portfolio"].plot(title="Portfolio Daily Returns")
plt.show()

# Histogram with KDE
plt.figure(figsize=(10, 4))
sns.histplot(df["Portfolio"], kde=True)
plt.title("Distribution of Portfolio Returns")
plt.show()

# Q-Q plot
stats.probplot(df["Portfolio"], dist="norm", plot=plt)
plt.title("Q-Q Plot of Portfolio Returns")
plt.grid(True)
plt.show()
```

## 4. Hypothesis Testing

I check:
- Whether portfolio return is statistically different from zero
- Whether it outperforms the benchmark (paired t-test)


```python
t_stat, p_val = stats.ttest_1samp(df["Portfolio"], 0)
print(f"T-test: t = {t_stat:.4f}, p = {p_val:.4f}")

if p_val < 0.05:
    print("Daily return is significantly different from 0.")
else:
    print("No significant difference from 0.")

t_stat_rel, p_val_rel = stats.ttest_rel(df["Portfolio"], df["Benchmark"])
print(f"Paired T-test: t = {t_stat_rel:.4f}, p = {p_val_rel:.4f}")

if p_val_rel < 0.05:
    print("Portfolio is significantly different from benchmark.")
else:
    print("No significant difference between portfolio and benchmark.")
```

## 5. Performance Metric

I calculate the Sharpe Ratio (mean return divided by standard deviation).


```python
sharpe_ratio = df["Portfolio"].mean() / df["Portfolio"].std()
print(f"Sharpe Ratio: {sharpe_ratio:.4f}")
```

## 6. Period-Based Hypothesis Testing

I split the dataset into two halves (before and after the midpoint) and:
- Run t-tests on each period
- Compare Sharpe Ratios
- Test against benchmark in each period


```python
midpoint = df.index[int(len(df)/2)]
before = df.loc[df.index < midpoint]
after = df.loc[df.index >= midpoint]

t_before, p_before = stats.ttest_1samp(before["Portfolio"], 0)
t_after, p_after = stats.ttest_1samp(after["Portfolio"], 0)

sharpe_before = before["Portfolio"].mean() / before["Portfolio"].std()
sharpe_after = after["Portfolio"].mean() / after["Portfolio"].std()

t_rel_before, p_rel_before = stats.ttest_rel(before["Portfolio"], before["Benchmark"])
t_rel_after, p_rel_after = stats.ttest_rel(after["Portfolio"], after["Benchmark"])

summary_df = pd.DataFrame({
    "Period": ["Before", "After"],
    "Mean Return t-stat": [t_before, t_after],
    "Mean Return p-val": [p_before, p_after],
    "Sharpe Ratio": [sharpe_before, sharpe_after],
    "Portfolio vs Benchmark t-stat": [t_rel_before, t_rel_after],
    "Portfolio vs Benchmark p-val": [p_rel_before, p_rel_after],
})
print(summary_df)
```

## 7. Logistic Regression Model

We predict whether the return is positive (Direction = 1) based on:
- Lagged return
- 5-day volatility

We evaluate the classification performance.


```python
features = df[["Lag_1", "Volatility_5d"]]
target = df["Direction"]

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## 8. Conclusion

- I successfully engineered time-based features (lag, volatility).
- The portfolio's daily return is statistically different from 0, but not significantly better than the benchmark.
- The Sharpe ratio is relatively low in both periods.
- Logistic Regression was used to classify return direction based on past data.
